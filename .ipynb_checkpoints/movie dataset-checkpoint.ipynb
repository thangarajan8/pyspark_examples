```python
def split_to_columns(text):
    return text.split("::")
```

```python
user_rows = usersrdd.map(split_to_columns)
```

```python
user_df = spark.createDataFrame(user_rows,schema=['user_id','gender','age','job','pin'])
```

```python
rating_df = spark.createDataFrame(ratings_rdd.map(lambda x:x.split("::")),schema=['user_id','movie_id','rating','timestamp'])
```

```python
movie_df = spark.createDataFrame(movies_rdd.map(lambda x:x.split("::")),schema=['movie_id','title','genre'])
```

```python
user_df.join(rating_df,on="user_id").collect()
```
